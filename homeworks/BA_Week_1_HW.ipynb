# Q5: 
## Write a function that takes a list and returns a new list that has all the duplicates removed.

Example input and expected output:
- input = `["Michele", "Robin", "Sara", "Michele"]`
- expected output = `['Michele', 'Robin', 'Sara']`


```python
# Solve problem here:

names = ["Michele", "Robin", "Sara", "Michele"]

def no_dups(lst:list[str]) -> list[str]:
    temp_unique_names = []
    for name in lst:
        if not name in temp_unique_names:
            temp_unique_names.append(name)
    return temp_unique_names

print(no_dups(names))
```

# Q6: Write a function that takes a list of numbers 
(for example, `a = [5, 10, 15, 20, 25]`) and returns a new list of only the first and last elements of the given list.

Example input and expected output:
- input = `[5, 10, 15, 20, 25]`
- expected output = `[5, 25]`

```python
# Solve problem here:
input_list = [5, 10, 99, 20, 25]

def index_elements_list(lst:list[int]) -> list[int]:
    temp_list:list[int] = []
    desired_indexes = [0, len(temp_list)-1]
    for i in desired_indexes:
        element = lst[i]
        temp_list.append(element)
    return temp_list

numbers = index_elements_list(input_list)
print(numbers)
```

# Q7: 
## Implement a function that takes as input three variables, and returns the largest of the three. 
### Try doing this without using the `max()` function!

_**Note:** all three input numbers will always be different, no need to account for a tie._

Example input and expected output:
- input: `your_function(1, 5, 10)`
- expected output:  `10`

```python
# Solve Problem here:

def my_max(a, b, c):
    # Fill in your code below and return max value of a, b, c
    nums = [a,b,c]
    temp_max = nums[0]
    for i in range(len(nums)-1):
        i2 = nums[i+1]
        if temp_max < i2:
            temp_max = i2
    return temp_max

max = my_max(1,5,10)
print(max)
    
```

# Q8: Write a function that takes a number as input and returns the following:
* If the input is divisible by three, return `'fizz'`
* If the input is divisible by five, return `'buzz'`
* If the input is divisible by three and by five, return `'fizzbuzz'`
* If the input is not divisible by three or five, return `None`.

```python
# Solve Problem fizzbuzz here:

def fizz_buzz(x:int):
    case_0 = (x % 3 == 0)
    case_1 = (x % 5 == 0)

    if case_0 and case_1:
        return "fizzbuzz"
    elif case_0:
        return "fizz"
    elif case_1:
        return "buzz"

    return None

print(fizz_buzz(15))
```

# Q2:  Covert all the rapper names to title case and save them into a new different list. 
Example: **lil wayne** becomes **Lil Wayne**

```python
# Solve rapper names here
rappers = ['lil wayne', 'nicki minaj', 'drake']

title_case_rappers = []

for rapper in rappers:
    new_rapper = ""
    
    for i in range(len(rapper)):
        c = rapper[i]
        if i == 0 and c.islower():
            new_rapper += c.upper()
        else:
            new_rapper += c

    title_case_rappers.append(new_rapper)
print(title_case_rappers)

```

# Q6: Write a function that takes a list of numbers 
(for example, `a = [5, 10, 15, 20, 25]`) and returns a new list of only the first and last elements of the given list.

Example input and expected output:
- input = `[5, 10, 15, 20, 25]`
- expected output = `[5, 25]`

```python
# Solve problem here:
input_list = [5, 10, 99, 20, 25]

def first_last(lst : list[int]) -> list[int]:
    if len(lst) == 0:
        return []
        
    if len(lst) == 1:
        return [lst[0]]
    
    fl = []
    fl.append(lst[0])
    fl.append(lst[-1])
    
    return fl

print(first_last(input_list))

test_case2 = [4]

print(first_last(test_case2))

```
